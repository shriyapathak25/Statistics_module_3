1. Hypothesis Testing
Hypothesis testing is a statistical method to determine if a hypothesis about a population is true or not, based on a sample of data.


2. Null and Alternative Hypotheses
- Null Hypothesis (H0): A statement that there is no significant difference or relationship.
- Alternative Hypothesis (H1): A statement that there is a significant difference or relationship.


3. Significance Level
The significance level (α) is the maximum probability of rejecting the null hypothesis when it is true. It's usually set to 0.05.


4. P-value
The P-value represents the probability of observing the sample data (or more extreme) assuming the null hypothesis is true.


5. Interpreting P-value
- If P-value < α (usually 0.05), reject the null hypothesis (significant result).
- If P-value ≥ α, fail to reject the null hypothesis (not significant)

6. Type 1 and Type 2 Errors
- Type 1 Error: Rejecting the null hypothesis when it's true (false positive).
- Type 2 Error: Failing to reject the null hypothesis when it's false (false negative).


7. One-Tailed vs Two-Tailed Test
- One-Tailed Test: Tests for an effect in one direction (e.g., increase or decrease).
- Two-Tailed Test: Tests for an effect in both directions (e.g., increase or decrease).


8. Z-Test
The Z-test is a statistical test used to compare a sample mean to a known population mean, assuming the population standard deviation is known.


9. Z-Score Calculation and Interpretation
- Calculate Z-score: (Sample Mean - Population Mean) / (Population Standard Deviation / √Sample Size)
- Z-score represents the number of standard deviations between the sample mean and the population mean.


10. T-Distribution
The T-distribution is used when the population standard deviation is unknown, and the sample size is small (usually < 30). It's used instead of the normal distribution to account for the uncertainty in estimating the population standard deviation.

11. Z-Test vs T-Test
- Z-Test: Used when population standard deviation is known.
- T-Test: Used when population standard deviation is unknown.


12. T-Test
The T-test is a statistical test used to compare a sample mean to a known population mean, or to compare the means of two samples, when the population standard deviation is unknown.

13. Z-Test and T-Test Relationship
The Z-test and T-test are both used for hypothesis testing, but the Z-test assumes a known population standard deviation, while the T-test estimates it from the sample data. As sample size increases, the T-test approaches the Z-test.


14. Confidence Interval
A confidence interval is a range of values within which a population parameter is likely to lie. It's used to interpret statistical results and estimate the population parameter.


15. Margin of Error
The margin of error is the maximum amount by which the sample estimate may differ from the true population parameter. A smaller margin of error means a narrower confidence interval, indicating more precision in the estimate.

16. Bayes' Theorem in Statistics
Bayes' Theorem is used to update probabilities based on new data. It helps statisticians make informed decisions by combining prior knowledge with new evidence

17. Chi-square Distribution
The Chi-square distribution is a statistical distribution used to model the sum of squared standard normal variables. It's commonly used for hypothesis testing and confidence intervals.


18. Chi-square Goodness of Fit Test
The Chi-square goodness of fit test is used to determine if observed data fit expected distributions. It compares observed frequencies to expected frequencies to assess the fit

19. F-Distribution
The F-distribution is a statistical distribution used to compare variances between two groups. It's commonly used in hypothesis testing, especially in ANOVA and regression analysis.


20. ANOVA Test
ANOVA (Analysis of Variance) is a statistical test used to compare means between multiple groups. Its assumptions include:
    - Normality of residuals
    - Equal variances between groups
    - Independence of observations

21. There are several types of ANOVA tests, including:

1-- One-Way ANOVA: Compares means between multiple groups.
2-- Two-Way ANOVA: Compares means between multiple groups with two independent variables.
3-- Repeated Measures ANOVA: Compares means between multiple groups with repeated measurements.
4-- Mixed Design ANOVA: Combines between-subjects and within-subjects designs.
5-- Multivariate ANOVA (MANOVA): Compares means between multiple groups with multiple dependent variables.


22.
1-- Compare means: F-test is used in ANOVA to compare means between multiple groups.
2-- Compare variances: F-test is used to compare variances between two or more groups.
3-- Test regression models: F-test is used to test the significance of regression models.

In hypothesis testing, the F-test is used to determine whether the observed differences between groups are statistically significant. If the F-statistic is significant (i.e., p-value < α), the null hypothesis is rejected, indicating that the differences between groups are statistically significant.

```python
#ans 1
import numpy as np
from scipy import stats

# Sample data
sample_mean = 25
sample_std = 5
sample_size = 100

# Population parameters
population_mean = 20
population_std = 5

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Perform Z-test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#ans 3
def one_sample_z_test(sample_data, population_mean, population_std):
    # Calculate sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Perform Z-test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

# Example usage
sample_data = np.array([23, 21, 19, 24, 22, 20, 18, 25, 23, 21])
population_mean = 20
population_std = 5

z_score, p_value = one_sample_z_test(sample_data, population_mean, population_std)

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

```

```python
#ans 5
def calculate_and_visualize_errors(alpha, beta, mu0, mu1, sigma):

    # Calculate critical value for Type 1 error
    critical_value = stats.norm.ppf(1 - alpha)

    # Calculate critical value for Type 2 error
    critical_value_beta = stats.norm.ppf(beta)
    # Visualize Type 1 error
    x = np.linspace(mu0 - 3 * sigma, mu0 + 3 * sigma, 100)
    plt.plot(x, stats.norm.pdf(x, loc=mu0, scale=sigma), label="Null distribution")
    plt.fill_between(x, stats.norm.pdf(x, loc=mu0, scale=sigma), where=(x > critical_value), color="red", alpha=0.5, label="Type 1 error")
    plt.legend()
    plt.title("Type 1 Error")
    plt.show()

    # Visualize Type 2 error
    x = np.linspace(mu1 - 3 * sigma, mu1 + 3 * sigma, 100)
    plt.plot(x, stats.norm.pdf(x, loc=mu1, scale=sigma), label="Alternative distribution")
    plt.fill_between(x, stats.norm.pdf(x, loc=mu1, scale=sigma), where=(x < critical_value_beta), color="blue", alpha=0.5, label="Type 2 error")
    plt.legend()
    plt.title("Type 2 Error")
    plt.show()

# Example usage
alpha = 0.05  # Significance level
beta = 0.2  # Probability of Type 2 error
mu0 = 0  # Population mean under the null hypothesis
mu1 = 1  # Population mean under the alternative hypothesis
sigma = 1  # Population standard deviation

calculate_and_visualize_errors(alpha, beta, mu0, mu1, sigma)
```

```python
#ans 9
def confidence_interval(sample, confidence_level=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample)
    sample_size = len(sample)
    standard_error = sample_std / np.sqrt(sample_size)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)
    margin_of_error = critical_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

# Sample data
np.random.seed(0)
sample = np.random.normal(loc=20, scale=5, size=100)

# Calculate confidence interval
lower_bound, upper_bound = confidence_interval(sample)

print(f"Sample mean: {np.mean(sample)}")
print(f"95% Confidence interval: ({lower_bound}, {upper_bound})")
```

```python
# ans 10
def calculate_margin_of_error(sample, confidence_level):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample)
    sample_size = len(sample)
    standard_error = sample_std / np.sqrt(sample_size)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)
    margin_of_error = critical_value * standard_error
    return margin_of_error

# Sample data
np.random.seed(0)
sample = np.random.normal(loc=20, scale=5, size=100)

# Calculate margin of error
confidence_level = 0.95
margin_of_error = calculate_margin_of_error(sample, confidence_level)

print(f"Sample mean: {np.mean(sample)}")
print(f"Margin of error for {confidence_level*100}% confidence level: {margin_of_error}")
```

```python
# ans 13
def calculate_expected_frequencies(observed_frequencies):
    row_sums = np.sum(observed_frequencies, axis=1)
    column_sums = np.sum(observed_frequencies, axis=0)
    total_sum = np.sum(observed_frequencies)

    expected_frequencies = np.outer(row_sums, column_sums) / total_sum
    return expected_frequencies

# Observed frequencies
observed_frequencies = np.array([[10, 20], [20, 30]])

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_frequencies)

print(f"Expected frequencies:\n{expected_frequencies}")
```

```python
# ans 19
def check_anova_assumptions(data):
    # Normality assumption
    for group in data:
        stat, p = stats.shapiro(group)
        if p < 0.05:
            print("Normality assumption not met")

    # Independence assumption
    # (Difficult to check programmatically, but can be assumed if data collection process is random)

    # Equal variance assumption
    f_statistic, p_value = stats.bartlett(*data)
    if p_value < 0.05:
        print("Equal variance assumption not met")

# Sample data
np.random.seed(0)
group1 = np.random.normal(loc=10, scale=2, size=100)
group2 = np.random.normal(loc=15, scale=3, size=100)
group3 = np.random.normal(loc=20, scale=4, size=100)

check_anova_assumptions([group1, group2, group3])
```

```python
#ans 20
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.factorplots import interaction_plot

# Step 1: Simulate sample data
np.random.seed(42)

# Independent variables (factors)
factor_A = np.repeat(['Low', 'Medium', 'High'], 10)  # First categorical factor (e.g., Treatment levels)
factor_B = np.tile(['A', 'B'], 15)  # Second categorical factor (e.g., Different groups)

# Dependent variable (response)
data_values = (np.random.normal(50, 10, 30) +  # Base value
               [5 if a == 'High' else -5 if a == 'Low' else 0 for a in factor_A] +
               [3 if b == 'B' else -3 for b in factor_B])

# Create a DataFrame
df = pd.DataFrame({'Factor_A': factor_A, 'Factor_B': factor_B, 'Value': data_values})

# Step 2: Perform Two-Way ANOVA
model = ols('Value ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)  # Type 2 ANOVA Table

# Step 3: Display ANOVA results
print("\nTwo-Way ANOVA Table:")
print(anova_table)

# Step 4: Visualize the results
plt.figure(figsize=(12, 5))

# Boxplot for factor A and B
plt.subplot(1, 2, 1)
sns.boxplot(x='Factor_A', y='Value', hue='Factor_B', data=df, palette='Set2')
plt.title("Boxplot of Values by Factors")

# Interaction plot
plt.subplot(1, 2, 2)
interaction_plot(x=df['Factor_A'], trace=df['Factor_B'], response=df['Value'], colors=['red', 'blue'])
plt.title("Interaction Plot")

plt.tight_layout()
plt.show()
```
